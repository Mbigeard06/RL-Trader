```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/blueChips/IBM.csv")
```

```python
log_return = np.log(df["close"] / df["close"].shift(1))
```

```python
hist = plt.hist(log_return, bins=100)
```

```python
mean = log_return.mean()
std = log_return.std()

print(f"mean :{mean}")
print(f"std :{std}")
```

```python
z_score = (log_return - mean) / std
z3 = np.abs(z_score) > 3
extreme_val = pd.Series(log_return[z3])
```

```python
z3_prop = extreme_val.count() / log_return.count()
print(f"Z > z3 percentage: {z3_prop*np.pow(10,2)}%") 
```

```python
extreme_Val = plt.hist(extreme_val, bins= 50)
```

```python
log_max = 4 * std + mean
n_to_clip = extreme_val[log_return > log_max].count()
n_to_clip
```
